# Text Issues

When classifying a document, a common word like "the" does nothing to inform us about the type of contents the document contains. But a word like "retenopathy" probably informs us that it's a medical document. The rarest words tend to teach us the most about a document. This is a huge problem for deep learning, since we need tons of training examples to build classifiers. 

Another problem is that two words may be different, but mean the same thing in certain constants. Cat and kitty can both be used to refer to feline animals in most cases, so we'd like to share weights most of the time, but once age becomes an issue, they no longer mean the same thing, and we'd like to not share the weights.

# Embedding

By using unsupervised learning, we can generate embeddings that predict context for words. We no longer have to even define words with unsupervised learning, their meaning will be inferred from context.

Word2Vec is a way to transform words into vector embeddings, and then use that vector to predict the next word. Then we train the same way we would as supervised learning.

##### Backprop Through Time
Stocastic Gradient Descent likes weights that do not depend on one another and are not correlated. This results in two problems:

1) The gradient "explodes" to infinity
2) The gradient "vanishes" to zero, and you end up not training anything

##### Fixing the Exploding Gradient and Vanishing Gradient
Exploding is a simple fix. We simply use a hack called *gradient clipping*. We basically take a weight maximum, and don't allow the weights to go beyond that point. It's hacky, but it's cheap and it is reasonably effective. 

Vanishing Gradient is a slightly more difficult fix that requires us to address the model itself. The solution to this the LSTM.
